### Practical tips for time series data science:

Time-series data can be tricky to work with because one, individual 'time-series' can sometime be unique which may limit the statistical inferences one can make. For instance if a researcher wants to forecast the stock marker in 2025 they can train a model on the years 2019 - 20223 to predict 2024 as a test. However the years 2019 - 2024 have a unique auto-correlation structure and in 2019 the pandemic happened. Then, that person can train a model on 2020 - 2024 to forecast 2025 and the weights that worked for the previous year may be inaccurate now. While this can make it difficult to accurately estimate metrics like standard errors a data scientist should keep two things in mind when drawing conculsions about the relationship between two variables[**1**].  

1. Consider whether relationships you find in the data are the result of a broader phenomenon or random variation in data.
2. If a phenomena, then consider how data sample might be representative of population that will be the forecasting target.

When it is appropriate to make stastical inferences with true replications it is also important to keep in mind the metrics that are used and what they mean. Paul the Octopus is a clairvoyent cephelapod that made soccer outcome predictions based on picking a colored box. Of course Paul has no idea what soccer is or how to evaluate the relative skill sets of two teams but he was correct 12/14 times. Under the null hypothesis that Paul is just selecting at random, this would result in a p-value of 0.0006 which is the same as saying one might say that this results would be expected 0.06% of time[**2**]. A very small number but not 0. In the world of data science it is sometimes possible to test the importantance of 1000 different features which means that 6 of those may be false positives such as the case of Paul.

Write down the AR($p$) model (mathematically) and show how to convert it into matrix equations for use with the one step ahead method. Put your final equation in the form $ax=b$, where $x$ is what you are solving for, $a$ is the matrix of coefficients (from the data) and $b$ is the dependent variable (the target you want to predict using the data). Note that the shape of these objects will depend on your choice of $p$, which you will want to vary.

$$
b _t = a_1 x_{t-1} + a_2 x_{t-2} + \cdots + a_p x_{t-p} + \epsilon_t
$$

$$
\begin{bmatrix} 
b_1\\
b_2\\
\end{bmatrix} = 
\begin{bmatrix}
x_1 & x_2 & \cdots & x_p \\
x_{2} & x_{3} & \cdots & x_{p+1}
\end{bmatrix}
\begin{bmatrix}
a_{p+1} \\
a_{p+2} \\
\end{bmatrix} +
\begin{bmatrix}
\epsilon_1 \\
\epsilon_{p+1} \\
\end{bmatrix}
$$

```python
import numpy as np
import pandas as pd

# center temps
df['Temp_scaled'] = (
    (df['Temperature'] - np.mean(df['Temperature']))
    / np.std(df['Temperature'])
)


# Assume `temperature_data` is the flattened time series
# Define AR order (p)
p = 3  # Number of lags

# Create lagged matrix A and target vector b
def create_ar_matrix(data, p):
    # Ensure data is a NumPy array
    data = np.array(data)
    
    # Create the lag matrix A
    A = np.column_stack([data[i:len(data)-p+i] for i in range(p)])
    
    # Create the target vector b
    b = data[p:]
    
    return A, b

# Generate A and b
A, b = create_ar_matrix(temperature_data, p)

# Print A and b shapes
print("Shape of A (lag matrix):", A.shape)
print("Shape of b (target vector):", b.shape)

# Example: Solve for AR coefficients using np.linalg.solve
# A.T @ A should be invertible for AR coefficients
x = np.linalg.solve(A.T @ A, A.T @ b)

print("AR Coefficients:", x)

# Step 1: Split into training and test sets
p = 3
train = df.loc[:len(df) - 60, :]
test = df.loc[len(df) - 60: , :]
test_ind = test['Date'][p:]

# create lagged matrix of train
A_train, b_train = create_ar_matrix(train['Temp_scaled'], p)
x = np.linalg.solve(A_train.T @ A_train, A_train.T @ b_train)
print("Train AR Coefficients:", x)

# make predictions
A_test, b_test = create_ar_matrix(test['Temp_scaled'], p)
pred = A_test @ x
df_pred = pd.DataFrame({"Date": test_ind, "Temp_scaled": np.array(pred)})

# Plot predictions
plt.figure(figsize=(12, 6))
plt.plot(train['Date'], train['Temp_scaled'], label="Training Data")
plt.plot(test['Date'], test['Temp_scaled'], label="Test Data", marker = 'o')
plt.plot(df_pred['Date'], df_pred['Temp_scaled'], label="Predictions", marker = 'x')
plt.title('Number of lags: {}'.format(p))
plt.legend()
plt.show()

# Step 1: Split into training and test sets
p = 12
train = df.loc[:len(df) - 60 , :]
test = df.loc[len(df) - 60: , :]
test_ind = test['Date'][p:]

# create lagged matrix of train
A_train, b_train = create_ar_matrix(train['Temp_scaled'], p)
x = np.linalg.solve(A_train.T @ A_train, A_train.T @ b_train)
print("Train AR Coefficients:", x)

# make predictions
A_test, b_test = create_ar_matrix(test['Temp_scaled'], p)
pred = A_test @ x
df_pred = pd.DataFrame({"Date": test_ind, 'Temp_scaled': np.array(pred)})

# Plot predictions
plt.figure(figsize=(12, 6))
plt.plot(train['Date'], train['Temp_scaled'], label="Training Data")
plt.plot(test['Date'], test['Temp_scaled'], label="Test Data", marker = 'o')
plt.plot(df_pred['Date'], df_pred['Temp_scaled'], label="Predictions", marker = 'x')
plt.title('Number of lags: {}'.format(p))
plt.legend()
plt.show()

# Step 1: Split into training and test sets
p = 1
train = df.loc[:len(df) - 60, :]
test = df.loc[len(df) - 60: , :]
test_ind = test['Date'][p:]

# create lagged matrix of train
A_train, b_train = create_ar_matrix(train['Temp_scaled'], p)
x = np.linalg.solve(A_train.T @ A_train, A_train.T @ b_train)
print("Train AR Coefficients:", x)

# make predictions
A_test, b_test = create_ar_matrix(test['Temp_scaled'], p)
pred = A_test @ x
df_pred = pd.DataFrame({"Date": test_ind, 'Temp_scaled': np.array(pred)})

# Plot predictions
plt.figure(figsize=(12, 6))
plt.plot(train['Date'], train['Temp_scaled'], label="Training Data")
plt.plot(test['Date'], test['Temp_scaled'], label="Test Data", marker = 'o')
plt.plot(df_pred['Date'], df_pred['Temp_scaled'], label="Predictions", marker = 'x')
plt.title('Number of lags: {}'.format(p))
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.vector_ar.var_model import VAR
from sklearn.preprocessing import StandardScaler

def train_and_predict_var(train_df, test_df, variables=['Temp', 'Precip'], p = 3):
    # Create copies to avoid modifying original data
    train = train_df.copy()
    test = test_df.copy()
    
    # Standardize the features
    scaler = StandardScaler()
    train_scaled = pd.DataFrame(
        scaler.fit_transform(train[variables]),
        columns=variables,
        index=train.index
    )
    test_scaled = pd.DataFrame(
        scaler.transform(test[variables]),
        columns=variables,
        index=test.index
    )
    
    # Train VAR model
    model = VAR(train_scaled)
    
    # Fit the model with the selected lag
    fitted_model = model.fit(p)
    print(f"Selected lag order: {p}")
    
    # Make predictions
    # We need lag_order number of observations from training data to start forecasting
    last_train_values = train_scaled.values[-p:]
    
    # Forecast for the length of test set
    forecast = fitted_model.forecast(y=last_train_values, steps=len(test))
    
    # Create predictions DataFrame
    predictions_scaled = pd.DataFrame(
        forecast,
        index=test.index,
        columns=variables
    )
    
    # Inverse transform the predictions to original scale
    predictions = pd.DataFrame(
        scaler.inverse_transform(predictions_scaled),
        index=test.index,
        columns=variables
    )
    
    return predictions, fitted_model
```
