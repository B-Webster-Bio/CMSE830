## Using Data Storytelling Frameworks
Remember the storytelling frameworks you just learned? They're perfect for organizing your presentation:

### ABCDE Approach for Your App
- **Analytics Type**: What kind of analysis does your app provide?
- **Background & Audience**: Who are your users? What do they need?
- **Context**: What problem does your app solve?
- **Data Narrative**: How does your solution work?
- **Execution**: How is it presented to users?

### 5W1H Framework for Technical Documentation
- **Why**: Why did you choose this problem?
- **Who**: Who benefits from your solution?
- **What**: What does your app do?
- **When**: When would users need this?
- **Where**: Where does the data come from?
- **How**: How does your solution work?

Discuss these with your team to ensure everyone is reminded of these frameworks. Incorporate this discussion into the tables you will create below.
