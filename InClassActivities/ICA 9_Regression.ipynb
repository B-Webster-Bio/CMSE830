<h3 style="font-family: 'Georgia', serif; color: #e74c3c;">Part 3: Use of the Loss Function in Linear Regression</h3>

Next, we'll explore how to handle the case of three or more data points mathematically using a _loss function_ and _optimization_. This function will allow us to measure how well the line fits the data and minimize the error. And, importantly, we will see that it works in nearly every case. 

When we fit a line to data using linear regression, we aim to minimize the error between the predicted values from our model and the actual data points. This error is captured by the _loss function_. This technique is extremely valuable in that it allows one to handle the case where there are many more data points than there are parameters in the model: this is the most common case in data science. 

The case in which there are fewer data points than parameters (the case above with two parameters $w_0$ and $w_1$ and one data point) can be handled; it is a bit subtle, so let's ignore it for now. 



<h4 style="font-family: 'Georgia', serif; color: #1a719a;">The Standard Loss Function</h4>


For linear regression, the most commonly used loss function is the **sum of squared errors (SSE)**:

$$ L(w_0, w_1) = \sum_{i=1}^{n} \left( y_i - (w_0 + w_1  x_i) \right)^2, $$

where:
- $i$ is the index for summing over all data points
- $n$ is the number of data points,
- $y_i$ are the actual values of the dependent variable,
- $w_0 + w_1 x_i$ is the predicted value from our model,
- $L(w_0, w_1)$ is the total error we want to minimize.

Minimizing the squared error gives us a best-fit line that balances the error across all points. The goal is to find the values of $w_0$ and $w_1$ that make the loss function as small as possible. In general, how many variables will the loss function depend on? 

<h4 style="font-family: 'Georgia', serif; color: #1a719a;">Partial Derivatives</h4>


To minimize the loss function, we take _partial derivatives_ with respect to $w_0$ and $w_1$ and set them to zero. This gives us the conditions that must be satisfied for the error to be minimized.

<h3 style="font-family: 'Georgia', serif; color: #e74c3c;">Part 4: Visualizing the Loss Function</h3>

Let's explore what these partial derivatives are doing. This can be visualized with the loss function $L(w_0,w_1)$ shown in the 2D space of parameters $(w_0,w_1)$ it depends on. Let's take a look at it. Be sure your group understands this code. 

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define a simple dataset with three points (x, y)
x_data = np.array([1, 2, 3])
y_data = np.array([2, 3, 5])

# Define the loss function (sum of squared errors)
def loss_function(w0, w1, x, y):
    return np.sum((y - (w0 + w1 * x))**2)

# Create a grid of w0 and w1 values for visualization
w0_vals = np.linspace(-5, 5, 100)
w1_vals = np.linspace(-5, 5, 100)
W0, W1 = np.meshgrid(w0_vals, w1_vals)

# Calculate the loss for each combination of w0 and w1
Z = np.array([[loss_function(w0, w1, x_data, y_data) for w1 in w1_vals] for w0 in w0_vals])

# Plot the loss function surface
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W0, W1, Z, cmap='viridis')
ax.set_title('Loss Function Surface: L(w0, w1)')
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('Loss')
plt.show()
```

When we find the minimum of this curve we are finding the optimal values of $w_0$ and $w_1$. At that minimum, the partial derivatives $\partial L/\partial w_i$ are both zero. That is, the loss function is **flat**: the derivatives in both directions are zero. 

```python
# Gradient descent function to minimize the loss function
def gradient_descent(x, y, w0_init, w1_init, learning_rate, num_iterations):
    w0, w1 = w0_init, w1_init
    m = len(y)
    for i in range(num_iterations):
        # Compute predictions
        y_pred = w0 + w1 * x
        # Calculate the gradients
        grad_w0 = (-2/m) * np.sum(y - y_pred)
        grad_w1 = (-2/m) * np.sum((y - y_pred) * x)
        # Update w0 and w1 using the gradients
        w0 -= learning_rate * grad_w0
        w1 -= learning_rate * grad_w1
        # Optionally print the progress
        if i % 100 == 0:
            print(f"Iteration {i}: w0 = {w0:.4f}, w1 = {w1:.4f}, loss = {loss_function(w0, w1, x, y):.4f}")
    return w0, w1

# Initial guess for w0 and w1
w0_init = 0
w1_init = 0
learning_rate = 0.01
num_iterations = 1000

# Perform gradient descent to find the optimal w0 and w1
optimal_w0, optimal_w1 = gradient_descent(x_data, y_data, w0_init, w1_init, learning_rate, num_iterations)

print(f"\nOptimal w0: {optimal_w0}, \nOptimal w1: {optimal_w1}")
```

As you can see, we can find the minimum of the loss function, which yields the "best" values of $w_0$ and $w_1$, using a numerical method that searches for the bottom. This might seem a bit more complicated than what you did at the beginning of this ICA where we just derived mathematical expressions. But, the numerical approach can be much better for complicated situations where we cannot derive simple expressions. 

Thus, you can see that partial derivatives have a very geometrical interpretation. 

Let's see if it worked:

```python
# Final plot showing the best-fit line along with the data points

# Generate a range of x values for the line
x_range = np.linspace(0, 4, 100)
y_pred = optimal_w0 + optimal_w1 * x_range

# Plot the three data points
plt.figure(figsize=(8, 6))
plt.scatter(x_data, y_data, color='red', label='Data Points')

# Plot the best-fit line
plt.plot(x_range, y_pred, color='blue', label=f'Best Fit Line: y = {optimal_w0:.2f} + {optimal_w1:.2f}x')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fitting Two Parameters to Three Data Points')
plt.legend()
plt.grid(True)
plt.show()
```

<h5 style="font-family: 'Georgia', serif; color: #FFA500;">Gradient Descent and Partial Derivatives</h5>


Weâ€™ve seen how **gradient descent** helps us minimize the loss function by iteratively updating the parameters $w_0$ and $w_1$. But how does this process actually work?

At each step of gradient descent, we compute the **partial derivatives** of the loss function with respect to $w_0$ and $w_1$. These partial derivatives tell us the direction in which the loss function is increasing. By moving in the opposite direction of the gradient, we ensure that the loss function decreases, allowing us to move closer to the minimum.

The loss function for linear regression is:

$$ L(w_0, w_1) = \sum_{i=1}^{n} \left( y_i - (w_0 + w_1 \cdot x_i) \right)^2 $$

To minimize this, we compute the partial derivatives:

$$ \frac{\partial L}{\partial w_0} = \frac{-2}{n} \sum_{i=1}^{n} \left( y_i - (w_0 + w_1 \cdot x_i) \right) $$

$$ \frac{\partial L}{\partial w_1} = \frac{-2}{n} \sum_{i=1}^{n} \left( y_i - (w_0 + w_1 \cdot x_i) \right) \cdot x_i .$$

Be absolutely sure everyone in your group can derive these and understands what they mean and are used for. Confirm below that everyone in your group studied this and understands it; if anyone is confused -- no problem -- write about what is confusing. 

These derivatives give us the gradient, which we use to update $w_0$ and $w_1$ (these are not real equal signs: they are like update equations you would have in Python):

$$ w_0 = w_0 - \alpha \cdot \frac{\partial L}{\partial w_0}, $$

$$ w_1 = w_1 - \alpha \cdot \frac{\partial L}{\partial w_1} ,$$

where $ \alpha $ is the learning rate that controls how large the update steps are.
